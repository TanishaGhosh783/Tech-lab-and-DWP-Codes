```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/harjotspahwa/Car-Evaluation/refs/heads/master/car_evaluation.csv')
print(df)
```

```python
col_names=['buying','maint','doors','persons','lug_boot','safety','class']
df.columns=col_names
col_names
#1. buying price
#2. maintance cost
#3. number of doors
#4. no.of persons
#5. lug_boot size
#6. safety
#7. class
```

```python
x=df.drop(['class'],axis=1)

y=df['class']
```

```python
#split x and y into training and testing sets
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test= train_test_split(x,y, test_size=0.33, random_state=42)
```

```python
!pip install category_encoders
import category_encoders as ce
```

```python
encoder=ce.OrdinalEncoder(cols=['buying','maint','doors','persons','lug_boot','safety'])

x_train=encoder.fit_transform(x_train)
x_test=encoder.transform(x_test)
```

```python
x_train
```

```python
from sklearn.tree import DecisionTreeClassifier
clf_gini=DecisionTreeClassifier(criterion='gini',max_depth=3,random_state=0)
clf_gini.fit(x_train,y_train)
```

```python
y_pred_gini=clf_gini.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
print('Model accuracy score with criterion gini index: {0:0.4f}'.format(accuracy_score(y_test,y_pred_gini)))
```

```python
y_pred_train_gini=clf_gini.predict(x_train)
y_pred_train_gini
```

```python
print('training set accuracy score: {0:0.4f}' . format(accuracy_score(y_train,y_pred_train_gini)))
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12,8))

from sklearn import tree
tree.plot_tree(clf_gini.fit(x_train,y_train))
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier(random_state=0)
rfc.fit(x_train,y_train)
y_pred=rfc.predict(x_test)
from sklearn.metrics import accuracy_score
print('Model accuracy score with criterion index: {0:0.4f}'.format(accuracy_score(y_test,y_pred)))
```
